## 3. ***Data Wrangling***

### Data Wrangling Code

####  Handling Missing Values

````python
# Handling Missing Values for CustomerID and Description

# 1. Dropping rows with missing CustomerID
# CustomerID is a critical field for identifying individual customers. Since it is a unique identifier,
# it cannot be meaningfully imputed. Retaining rows with missing CustomerIDs would distort customer-level analysis.
# Therefore, we drop rows where CustomerID is missing to maintain data integrity and ensure accurate segmentation.

df = df.dropna(subset=['CustomerID'])

# 2. Dropping rows with missing Description
# The 'Description' column provides vital information about the product. Imputing missing product descriptions
# could introduce inaccuracies, as each product has a specific description that is difficult to guess.
# Dropping rows with missing descriptions helps maintain product-related accuracy and ensures consistency in analysis.

df = df.dropna(subset=['Description'])

# Resetting the index after dropping rows to maintain a clean and orderly DataFrame.
# This is a good practice to avoid index misalignment after modifying the data.

df.reset_index(drop=True, inplace=True)

# Showing the refined dataset shape after handling missing values
print(f"After addressing the missing values, the dataset now contains {df.shape[0]} rows and {df.shape[1]} columns.")
````

####  Handling Duplicate Entries

````python
# Removing Duplicate Entries

# In the dataset, duplicate entries represent repeated records, which do not contribute additional useful information.
# Keeping these duplicate rows can inflate key metrics, distort customer segments, and negatively impact machine learning models.
# For example, repeated purchase records can cause inflated transaction counts and mislead the segmentation process.
# Therefore, it's crucial to remove duplicates to ensure that analysis and models are based on unique and accurate customer data.

# Dropping duplicate rows from the dataset
df = df.drop_duplicates()

# Resetting the index after removing duplicates to maintain data integrity and a clean DataFrame
df.reset_index(drop=True, inplace=True)

# Verifying the dataset shape after removing duplicates
print(f"Updated dataset shape after eliminating duplicate entries: {df.shape}")
````

####  Data Type Conversion

````python
# Convert 'InvoiceDate' from object to datetime format
# This will allow us to extract features like date, month, year, day of week etc.
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'],format='mixed', dayfirst=True)

# Convert 'CustomerID' from float to string
# Although it contains numbers, 'CustomerID' is a categorical identifier and not meant for mathematical operations
df['CustomerID'] = df['CustomerID'].astype(str)

# Verifying the updated data types
print("\nUpdated Data Types:\n")
print(df.dtypes)
````

####  Creating a New Feature

````python
# Creating a new feature 'TotalAmount'
# This column represents the total value of each transaction line
# and is crucial for analyzing customer spending behavior

df['TotalAmount'] = df['Quantity'] * df['UnitPrice']

# Verifying the new column
df[['Quantity', 'UnitPrice', 'TotalAmount']].head()
````

####  Filtering Invalid Transactions

````python
# Removing transactions where Quantity or UnitPrice is less than or equal to 0
# These are likely returns, cancellations, or incorrect entries

df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Resetting the index for a clean DataFrame
df.reset_index(drop=True, inplace=True)

# Confirming the shape after cleaning
print(f"Data shape after filtering invalid transactions: {df.shape}")
````

### What all manipulations have you done and insights you found?

The data was refined through several key steps:

* **Handling Missing Values**: Rows with missing CustomerID and Description were removed to maintain data integrity and ensure accurate analysis.

* **Removing Duplicates**: Duplicate entries were eliminated to keep only unique records, ensuring accurate customer segmentation.

* **Data Type Conversion**: The InvoiceDate was converted to a datetime format for time-based analysis, and CustomerID was changed to a string for correct categorical treatment.

* **Feature Engineering**: A new TotalAmount feature was created by multiplying Quantity and UnitPrice to better understand customer spending behavior.

* **Filtering Invalid Transaction**s: Rows with zero or negative values in Quantity or UnitPrice were filtered out to ensure data quality.

As a result, the dataset was reduced from 541,909 to 392,692 rows, with a new TotalAmount feature, making it cleaner and ready for meaningful analysis.

#### Chart - 1  Univariate Analysis

````python
# Chart - 1 visualization code
# Visualization Function - Top N Countries with the Most Orders
def plot_top_countries_distribution(df, top_n=15):
    """
    Plots the distribution of the top N countries based on number of transactions.

    """
    # Get count of transactions per country
    country_counts = df['Country'].value_counts().nlargest(top_n).reset_index()
    country_counts.columns = ['Country', 'OrderCount']

    # Initialize the plot
    plt.figure(figsize=(10, 5))
    ax = sns.barplot(
        data=country_counts,
        x='OrderCount',
        y='Country',
        hue='Country',
        palette='viridis',
        dodge=False,
        legend=False  # Hide legend since it's redundant with y-axis labels
    )

    # Add value labels to bars
    for i, v in enumerate(country_counts['OrderCount']):
        ax.text(v + 100, i, str(v), va='center', fontsize=11, color='black')

    # Customize appearance
    plt.title(f'Top {top_n} Countries with the Most Orders', fontsize=16)
    plt.xlabel('')  # Remove x-axis label
    plt.xticks([])  # Remove x-axis ticks
    plt.ylabel('Country', fontsize=12)
    sns.despine(left=True, bottom=True)

    plt.tight_layout()
    plt.show()
# Call the function to display top 10 countries
plot_top_countries_distribution(df, top_n=15)
````

````python
# Chart - 5 visualization code
# Function to plot the top N countries based on total spending
def plot_spending_by_country(df, top_n=10):
    """
    Plots the top N countries based on total spending using the 'TotalAmount' column.
    """
    # Get top N countries by total spending
    top_countries = df.groupby('Country')['TotalAmount'].sum().nlargest(top_n).reset_index()
    top_countries.columns = ['Country', 'TotalSpending']

    # Initialize the plot
    plt.figure(figsize=(10, 5))
    ax = sns.barplot(
        data=top_countries,
        x='TotalSpending',
        y='Country',
        hue='Country',
        palette='viridis',
        dodge=False,
        legend=False
    )

    # Add value labels to bars
    for i, v in enumerate(top_countries['TotalSpending']):
        ax.text(v + top_countries['TotalSpending'].max() * 0.01, i, f'{v:,.0f}', va='center', fontsize=11)

    # Customize appearance
    plt.title(f'Top {top_n} Countries by Total Spending', fontsize=16)
    plt.xlabel('')  # Remove x-axis label
    plt.xticks([])  # Remove x-axis ticks
    ax.xaxis.grid(False)  # Remove vertical grid lines
    plt.ylabel('Country', fontsize=12)
    sns.despine(left=True, bottom=True)  # Remove axis spines

    # Display the plot
    plt.tight_layout()
    plt.show()

# Call the function to plot the top 10 countries by total spending
plot_spending_by_country(df, top_n=10)
````

#### Chart - 7  Time Series Analysis

````python
# Function to visualize sales trends based on the selected plot type
def plot_sales_trends(df, plot_type='monthly_quantity'):
    """
    Visualizes different sales trends based on the selected plot type.

    Available plot types:
    - 'monthly_quantity': Line plot of total quantity sold per month.
    - 'daily_revenue': Line plot of total daily revenue over time.
    - 'monthly_revenue': Line plot of monthly revenue trends.
    """
    # Ensure 'InvoiceDate' is in datetime format
    df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

    # Create the plot based on the plot_type
    if plot_type == 'monthly_quantity':
        df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')
        monthly_quantity = df.groupby('InvoiceMonth')['Quantity'].sum()
        plt.figure(figsize=(7, 4))
        monthly_quantity.plot(marker='o', linestyle='-', color='#27408B')
        plt.title('Monthly Quantity of Items Sold', fontsize=16)
        plt.xlabel('Month', fontsize=12)
        plt.ylabel('Total Quantity Sold', fontsize=12)

    elif plot_type == 'daily_revenue':
        df['InvoiceDay'] = df['InvoiceDate'].dt.date
        daily_revenue = df.groupby('InvoiceDay')['TotalAmount'].sum()
        plt.figure(figsize=(7, 4))
        daily_revenue.plot(linestyle='-', color='#27408B')
        plt.title('Daily Revenue Over Time', fontsize=16)
        plt.xlabel('Date', fontsize=12)
        plt.ylabel('Total Revenue', fontsize=12)

    elif plot_type == 'monthly_revenue':
        df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')
        monthly_revenue = df.groupby('InvoiceMonth')['TotalAmount'].sum()
        plt.figure(figsize=(7, 4))
        monthly_revenue.plot(marker='o', linestyle='-', color='#27408B')
        plt.title('Monthly Revenue Trend', fontsize=16)
        plt.xlabel('Month', fontsize=12)
        plt.ylabel('Revenue', fontsize=12)

    else:
        print("Invalid plot type. Choose from 'monthly_quantity', 'daily_revenue', or 'monthly_revenue'.")
        return

    # Display the plot
    plt.tight_layout()
    plt.show()
````

````python
# Chart - 7 visualization code: Visualizing the Monthly Quantity of Items Sold
# Call the function to visualize the total quantity sold per month
plot_sales_trends(df, plot_type='monthly_quantity')
````

````python
# Selecting numerical columns relevant for outlier detection
numeric_cols = ['Quantity', 'UnitPrice', 'TotalAmount']

def plot_boxplots(df, cols):
    """
    Plots distributions with log scale transformation for better outlier visibility.
    """
    colors = sns.color_palette("Set2", len(cols))  # Cool-to-warm color palette for variety
    plt.figure(figsize=(15, 4))

    for i, col in enumerate(cols, 1):
        plt.subplot(1, len(cols), i)

        # Filter positive values and apply log1p for scale compression
        data = df[col][df[col] > 0]
        sns.boxplot(y=np.log1p(data), color=colors[i - 1])

        # Title as distribution instead of boxplot
        plt.title(f'Distribution of {col} (Log Scale)', fontsize=12)
        plt.ylabel('Log(1 + Value)')
        plt.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

# Call the function
plot_boxplots(df, numeric_cols)
````

The box plots on a log scale nicely pinpoint potential outliers in Quantity, UnitPrice, and TotalAmount. The points beyond the whiskers highlight values that stand out. More potentially high outliers are observed across these variables, necessitating careful consideration during dataset refinement.

````python
# Handling Outliers & Outlier treatments
# Function to calculate the Interquartile Range (IQR) for a specified column
def calculate_iqr(df, column):
    """
    Calculates the Interquartile Range (IQR) for a given column in the DataFrame.

    """
    Q1 = df[column].quantile(0.25)  # 25th percentile (Q1)
    Q3 = df[column].quantile(0.75)  # 75th percentile (Q3)

    IQR = Q3 - Q1  # Interquartile Range

    # Calculate lower and upper bounds for outlier detection
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return lower_bound, upper_bound


# Function to remove outliers from a specified column using the IQR method
def remove_outliers_iqr(df, column):
    """
    Removes outliers from a specified column using the IQR method.

    Outliers are defined as values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR.

    """
    # Get lower and upper bounds for outlier detection
    lower_bound, upper_bound = calculate_iqr(df, column)

    # Filter out rows with outliers
    df_cleaned = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    # Print the number of removed outliers
    num_outliers = df.shape[0] - df_cleaned.shape[0]
    print(f"Removed {num_outliers} outliers from '{column}'")

    return df_cleaned


# Function to clean the data by removing outliers from multiple columns
def clean_data(df, columns):
    """
    Cleans the DataFrame by removing outliers from specified columns.

    """
    for column in columns:
        df = remove_outliers_iqr(df, column)

    return df


# List of columns to remove outliers from
numeric_cols = ['Quantity', 'UnitPrice', 'TotalAmount']

# Apply the outlier removal process
df_cleaned = clean_data(df, numeric_cols)
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

The Interquartile Range (IQR) method was used for outlier detection and treatment.

How it works: Outliers are identified based on values outside the range Q11.5IQRQ11.5IQR to Q3+1.5IQRQ3+1.5IQR, where Q1 and Q3 are the first and third quartiles of the data.

**Reason for Use**:
* **Effective for skewed data**: IQR works well with non-normally distributed data.
* **Non-parametric**: It doesnt assume a specific data distribution, making it adaptable.
* **Simple and widely accepted**: Easy to implement and interpret, making it a reliable choice for outlier detection.

This method ensures outliers are removed without distorting the data's underlying structure.

#### 1. Feature Manipulation

````python
# Manipulate Features to minimize feature correlation and create new features
# 'Recency': Measures the number of days since the last transaction per customer.
# Function to calculate 'Recency' based on the latest InvoiceDate
def calculate_recency(df, reference_date):
    recency = df.groupby('CustomerID')['InvoiceDate'].max().reset_index()
    recency['Recency'] = (reference_date - recency['InvoiceDate']).dt.days
    return recency[['CustomerID', 'Recency']]

# Define the reference date and call the recency calculation function
reference_date = df['InvoiceDate'].max()
recency_df = calculate_recency(df, reference_date)
recency_df
````

````python
# 'Frequency' : Captures the number of unique purchase transactions per customer.
# Function to compute 'Frequency' based on count of unique invoices
def calculate_frequency(df):
    frequency = df.groupby('CustomerID')['InvoiceNo'].nunique().reset_index()
    frequency.columns = ['CustomerID', 'Frequency']
    return frequency

# Call the function to compute customer frequency
frequency_df = calculate_frequency(df)
frequency_df
````

````python
# 'Monetary' : Represents the total spending per customer.
# Function to calculate 'Monetary' as the sum of TotalAmount per customer
def calculate_monetary(df):
    monetary = df.groupby('CustomerID')['TotalAmount'].sum().reset_index()
    monetary.columns = ['CustomerID', 'Monetary']
    return monetary

# Call the function to compute monetary value
monetary_df = calculate_monetary(df)
monetary_df
````

````python
# Function to merge recency, frequency, and monetary datasets : Creates a consolidated dataset for customer segmentation.
def create_rfm(recency_df, frequency_df, monetary_df):
    rfm = recency_df.merge(frequency_df, on='CustomerID')
    rfm = rfm.merge(monetary_df, on='CustomerID')
    return rfm

# Merge RFM components to form a single dataset
rfm_df = create_rfm(recency_df, frequency_df, monetary_df)
rfm_df
````

````python
# 'Average Order Value' : Calculated as Monetary divided by Frequency.
# Function to compute average order value per customer
def add_avg_order_value(rfm_df):
    rfm_df['AvgOrderValue'] = rfm_df['Monetary'] / rfm_df['Frequency']
    return rfm_df

# Add the 'AvgOrderValue' column
rfm_df = add_avg_order_value(rfm_df)
rfm_df
````

````python
# 'Customer Tenure' : Measures how long a customer has been active since their first purchase.
# Function to calculate tenure in days since the customer's first purchase
def calculate_tenure(df, reference_date):
    tenure = df.groupby('CustomerID')['InvoiceDate'].min().reset_index()
    tenure['TenureDays'] = (reference_date - tenure['InvoiceDate']).dt.days
    return tenure[['CustomerID', 'TenureDays']]
# Compute and merge the customer tenure with RFM dataset
tenure_df = calculate_tenure(df, reference_date)
rfm_df = rfm_df.merge(tenure_df, on='CustomerID')
rfm_df
````

````python
# 'Order Rate' : Frequency normalized by tenure to understand purchasing rate.
# Function to compute order rate = Frequency / TenureDays
def add_order_rate(rfm_df):
    rfm_df['OrderRate'] = rfm_df['Frequency'] / rfm_df['TenureDays']
    return rfm_df

# Add the 'OrderRate' column
rfm_df = add_order_rate(rfm_df)
rfm_df
````

````python
# Train KMeans using the Elbow method
def elbow_method(X_scaled , k_range=(1, 15)):
    model = KMeans(init='k-means++', max_iter=1000, random_state=42)
    visualizer = KElbowVisualizer(model, k=k_range, metric='distortion', timings=False)
    visualizer.fit(X_scaled)
    visualizer.show()
    return visualizer.elbow_value_
````

````python
#  Silhouette Analysis for various cluster counts
def silhouette_analysis(X_scaled , max_k=7):
    for n_clusters in range(2, max_k):
        km = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
        preds = km.fit_predict(X_scaled )
        score = silhouette_score(X_scaled , preds)
        print(f'For n_clusters = {n_clusters}, silhouette score is {score:.4f}')

        visualizer = SilhouetteVisualizer(km, colors='Set1')
        visualizer.fit(X_scaled )
        visualizer.show()
````

The Silhouette Score shows that for k = 5, the score is 0.2882, indicating reasonable cluster separation. The score slightly improves to 0.3023 for k = 6, but then decreases for k = 7 and beyond, suggesting additional clusters dont significantly improve separation.

**Conclusion**: k = 5 is the ideal choice for clustering, balancing both inertia and cluster separation effectively.

````python
# Function: plot_clusters_pca
# Purpose: Train a KMeans clustering model and visualize the results in 2D space
# using Principal Component Analysis (PCA) for dimensionality reduction.
#  Define the plot_clusters_pca function to visualize KMeans results
def plot_clusters_pca(X_scaled, n_clusters):
    """
    Trains a KMeans model and visualizes the clusters in 2D using PCA.
    """
    # Fit KMeans and predict labels
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
    y_km = kmeans.fit_predict(X_scaled)

    # Apply PCA to reduce to 2D
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X_scaled)
    centers_pca = pca.transform(kmeans.cluster_centers_)

    # Plotting the clusters
    plt.figure(figsize=(10, 6))
    plt.title(f'Customer Segmentation (K={n_clusters}) using PCA')
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_km, cmap='Set3', s=50, alpha=0.7, label='Data Points')
    plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='black', s=200, marker='x', label='Centroids')

    # Annotate each cluster centroid with its cluster label
    for i, (x, y) in enumerate(centers_pca):
        plt.annotate(f'Cluster {i}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')

    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
````

````python
#  Final cluster visualization
plot_clusters_pca(X_scaled , n_clusters= 5)
````

````python
# Function to calculate k-distance graph
def calculate_k_distance(X_scaled, n_neighbors=5):
    """
    This function calculates the k-distance for each point in the dataset,
    which is used to identify the optimal epsilon value for DBSCAN.
    """
    # Fit the NearestNeighbors model to the scaled data
    neighbors = NearestNeighbors(n_neighbors=n_neighbors)
    neighbors_fit = neighbors.fit(X_scaled)

    # Get the distances and indices of the nearest neighbors
    distances, indices = neighbors_fit.kneighbors(X_scaled)

    # Sort the distances in ascending order for visualization
    distances = np.sort(distances[:, -1], axis=0)

    return distances

# Function to plot the k-distance graph
def plot_k_distance_graph(distances):
    """
    This function plots the k-distance graph to help determine the optimal epsilon
    value for DBSCAN. The 'elbow' in the graph indicates the ideal epsilon.
    """
    # Plot the sorted k-distances
    plt.plot(distances)
    plt.title('k-distance Graph')
    plt.xlabel('Points sorted by distance')
    plt.ylabel('Distance')
    plt.show()

# Calculate the k-distance and plot the graph
distances = calculate_k_distance(X_scaled)
plot_k_distance_graph(distances)
````

````python
# Function to fit DBSCAN with given hyperparameters
def fit_dbscan(X_scaled, eps_value, min_samples_value):
    """
    Fits the DBSCAN clustering model with specified epsilon and min_samples values.

    """
    # Fit the DBSCAN model with given eps and min_samples
    dbscan_model = DBSCAN(eps=eps_value, min_samples=min_samples_value, metric='euclidean')
    cluster_labels = dbscan_model.fit_predict(X_scaled)
    return cluster_labels

# Function to calculate silhouette score for the clustering results
def evaluate_silhouette_score(X_scaled, cluster_labels):
    """
    Calculates the silhouette score to evaluate the clustering performance.

    """
    return silhouette_score(X_scaled, cluster_labels)

# Step 2: Tune hyperparameters and evaluate performance

# Define hyperparameter ranges for DBSCAN
eps_values = [0.75, 0.8, 0.85, 0.9, 0.95]  # Example epsilon values
min_samples_values = [3, 5, 7]  # Example min_samples values

# Initialize variables to track the best hyperparameters and silhouette score
best_silhouette = -1
best_eps = None
best_min_samples = None

# Loop through all combinations of eps and min_samples values
for eps in eps_values:
    for min_samples in min_samples_values:
        # Fit DBSCAN with current hyperparameters
        cluster_labels = fit_dbscan(X_scaled, eps, min_samples)

        # Calculate silhouette score for the current clustering result
        silhouette_avg = evaluate_silhouette_score(X_scaled, cluster_labels)

        # Print the current hyperparameters and corresponding silhouette score
        print(f"eps: {eps}, min_samples: {min_samples} -> Silhouette Score: {silhouette_avg}")

        # Update best hyperparameters if current silhouette score is higher
        if silhouette_avg > best_silhouette:
            best_silhouette = silhouette_avg
            best_eps = eps
            best_min_samples = min_samples

# Output the best hyperparameters and silhouette score
print(f"\nBest eps: {best_eps}, Best min_samples: {best_min_samples} with Silhouette Score: {best_silhouette}")
````

We used a combination of k-distance graph and Grid Search for hyperparameter optimization. After determining a reasonable range for epsilon, we performed a Grid Search over different combinations of epsilon (0.75, 0.8, 0.85, 0.9, 0.95) and min_samples (3, 5, 7) to find the best clustering performance for DBSCAN.
